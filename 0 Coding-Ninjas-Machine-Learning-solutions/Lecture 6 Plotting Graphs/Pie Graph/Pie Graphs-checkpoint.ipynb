## Plot function x^3 

```python
import matplotlib.pyplot as plt
import numpy as np
#x=np.array([1,2,3,4])
x=np.arange(0,5,0.1)
#color,marker,line width for axis,dataset and title,axis and grid
y=x**3
y2=x**2
#plt.plot(x,y,color="black",marker="0)
plt.plot(x,y,color="black",linewidth=5,label="x^3")
plt.plot(x,y2,color="red",linewidth=5,label="x^2")
plt.axis([0,10,0,200])
plt.ylabel("y")
plt.xlabel("x")
plt.grid()
plt.text(2,80,"text",fontsize=12)
plt.legend()
plt.title("Matplotlib Demo")
plt.show()
```

## Pie Graph

```python
labels=["A","B","C","D"]
size=[3,4,6,2]
colors=["blue","purple","red","pink"]
explode=[0.1,0,0,0]
plt.title("Split among classes")
plt.pie(size,colors=colors,explode=explode,labels=labels,autopct="%.2f%%",counterclock=False,startangle=100)
plt.axis("equal")
plt.show()
```
