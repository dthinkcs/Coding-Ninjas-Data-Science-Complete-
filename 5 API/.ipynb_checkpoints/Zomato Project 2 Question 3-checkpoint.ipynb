### Question 3

#### 3.1 Number of occurences of each restaurant is counted and plotted on a bar graph for top 10

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('zomato.csv')

df['Restaurant Name'].value_counts().head(10).plot(kind='bar')
```

#### Question 3.2 NaN entries are removed

```python
df2 = df[df['Aggregate rating'].notnull()]
df2.hist(column = 'Aggregate rating')
```

#### Question 3.3 data is sorted based on number of votes per restaurant

```python
df3 = df.sort_values(['Votes'], ascending=[False]).head(10)
df3.plot(x='Restaurant Name', y='Votes', kind = 'bar')
```

#### Question 3.4
in the following code I first split the cuisines and made seperate row for each cuisine (line 5 to 15) Then I count the number of each cuisines present in datafram and sorted and took top 10

```python
df4 = df[df['Country Code']==216]
df4 = df[['Cuisines']]
df4 = df4[df4['Cuisines'].notnull()]

def foo(a):
    s = a.split(',')
    return s

df4['Cuisines'] = df4['Cuisines'].apply(lambda x: foo(x))

df4.reset_index(inplace=True)
rows = []
_ = df4.apply(lambda row: [rows.append([nn])for nn in row.Cuisines], axis=1)

df_new = pd.DataFrame(rows, columns=['cuisine'])

df_new.groupby('cuisine').size().sort_values(ascending=False).head(10).plot(kind='pie')
```

#### Question 3.5
 selected country as India
 
 A new column 'prod' is created which is the product of rating into votes 
 
 data is grouped by City and sum and count of votes and prod for each locality is calculated, the result is stored in 'a' 
 
a new column w is created which is prod/Votes

Scatter plot is show as required

```python
df5 = df[df['Country Code']==1]
df5['prod'] = df5['Aggregate rating'] * df5['Votes']

a = df5.groupby("City").agg(['sum','count'])
a = a[['Votes','prod']]
a['w'] = a['prod']['sum'] / a['Votes']['sum']
a = a[['w','Votes']] 
a.columns = a.columns.droplevel()
del a['sum']

a = a.reset_index(drop=False)

a.columns = ['City','w','count']


xi = np.array(a['City'])
yi = np.array(a['count'])
si = np.array(a['w'])

plt.scatter(x=xi,y=yi,s=si*10,marker='o')#, width=1000, height=1000)
plt.xticks(rotation=90)
plt.xlabel('City')
plt.ylabel('Number of restaurantsr')
plt.legend(['weighted restaurant rating'])

plt.show()
```
