## DATA Cleaning

4 : data is loaded
5 : Countries other than India are rejected
6 : Cities ["New Delhi", "Ghaziabad", "Noida", "Gurgaon", "Faridabad"] are replaced by Delhi NCR for further         analysis

```python
import pandas as pd
import numpy as np

df = pd.read_csv('zomato.csv')
df = df[df['Country Code'] == 1]
df['City'].replace(to_replace =["New Delhi", "Ghaziabad", "Noida", "Gurgaon", "Faridabad"], value ="Delhi NCR", inplace=True) 

df
```

### Question 1.1
2 : Number of restaurants in Delhi NCR is calculated
3 : Number of restaurants not in Delhi NCR is calculated
8 : A graph is plotted between the two

```python
print(df.City.count())
print(df[df.City == 'Delhi NCR'].City.count())
print(df[df.City != 'Delhi NCR'].City.count())

import matplotlib.pyplot as plt
%matplotlib inline

plt.bar(['Delhi NCR', 'Rest of India'], [7947,705])
plt.ylabel("No. of restraunts")
plt.show()
```

### Question 1.2 

1-5 : Data type of cuisines is string, hence it is first split and converted to a set of cuisines seperatly for Delhi NCR and other cities

        set1 : set of cuisines served in restaurants in delhi ncr
        set2 : set of cuisines served in restaurants in rest of India
        
7 : set of cuisines served in rest of India but not in delhi is printed 
10-20 : A request is made to zomato API and a new_set of cuisines is created for Delhi NCR (city id = 1)
22-23 : The set of elements printed in line 7 are checked for presense in new_set. If true, then incomplete dataset, if false then cuisine is actually not served in delhi ncr

```python
d1 = df[df['City']=='Delhi NCR']['Cuisines'].str.split(', ')
set1 = set().union(*d1)

d2 = df[df['City'] !='Delhi NCR']['Cuisines'].str.split(', ')
set2 = set().union(*d2)

print(set2 - set1)


import requests

h = {'user-key' : 'a6e0142cda800d18ad52d65649840fcd'}
p = {'city_id' :1}

r = requests.get('https://developers.zomato.com/api/v2.1/cuisines', headers = h, params = p)

new_set = set([]) 

for i in r.json()['cuisines'] :
    new_set.add(i['cuisine']['cuisine_name']) 
    
for i in set2-set1 :
    print(i, i in new_set)
```

### Question 1.3

1-5 : A dictionary is created for cuisines served in Delhi NCR. Key = cuisine, Value = No. of restaurants that serve that cuisine

7-16 : arrays x and y are populated to plot graph
18-20 : required graph is plotted

```python
df1 = pd.DataFrame(d1) # d1 is defined in previous block
dic1 = {}
for index, row in df1.iterrows():
    for i in row['Cuisines']:
        dic1[i] = dic1.get(i,0) + 1
        
count = 0   
x = []
y = []
for w in sorted(dic1, key=dic1.get, reverse=True):
    x.append(w)
    y.append(dic1[w])
    print (w, dic1[w])
    count += 1
    if count == 10 :
        break
        
plt.bar(x, y, align='center')
plt.xticks(range(len(x)), list(x), rotation = 30)
plt.show()
```
