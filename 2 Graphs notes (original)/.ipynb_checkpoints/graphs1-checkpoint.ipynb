```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [2,4,6]

plt.scatter(x, y)
plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]

x2 = [3,4,5]
y2 = [2,4,6]

plt.plot(x2, y2, "b-.")

plt.plot(x, y)
plt.show()
```

```python
# plot function x^3
import numpy as np
#x = np.array([1,2,3,4])
x = np.arange(0, 5, 0.1)
y  = x **3
plt.plot(x, y, "b")
plt.show()
```
