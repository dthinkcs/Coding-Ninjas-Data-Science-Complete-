## Question 1 :  Login
>Login to your Instagram Handle. Submit with sample username and password


- <font color=red> Login process is done only till username,password submission since, after that sometimes there is two factor authorisation and sometimes it asks if you want to enable notifications. That has to be done manually </font>

```python
def login(driver,usr,pswd) :
    
    try :
        reg_btn = wait.until(EC.presence_of_element_located((By.LINK_TEXT,'Log in')))
        reg_btn.click()
        time.sleep(2) 
        text_box = wait.until(EC.presence_of_element_located((By.NAME,'username')))
        text_box.send_keys(usr)
        text_box = driver.find_element_by_name('password')
        text_box.send_keys(pswd)
        text_box.submit()
    
        #uncomment the line below if want to click "Not now" in notification alert
        #wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, ".aOOlW.HoLwm"))).click()
    
    except TimeoutException :
        print ("Can't Login")
```

## Question 2 : Create List

>Type for ‚Äúfood‚Äù in search bar and print all the names of the Instagram Handles that are displayed in list after typing ‚Äúfood‚Äù. Make sure to avoid printing hashtags

- <font color=red> All options except hashtags are printed. </font>

```python
def create_list(driver, query) :
    try :
        
        #Locating the search bar, clearing it, and sending the query
        search_box = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'XTCLo')))
        search_box.clear()
        search_box.send_keys(query)
        
        #wait till elements are loaded
        query = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'yCE8d')))
        
        #make a list of all loaded elements
        li = driver.find_elements_by_class_name('yCE8d')
        final_list = []
        
        #extracting the elements which are not starting with'#'
        for i in li :
            if i.text[0] != '#' :
                final_list.append(i.text.split('\n')[0])             
        
        #closing the search bar
        try : 
            driver.find_element_by_class_name('coreSpriteSearchClear').click()
        except Exception as e:    
            pass
        
        #return the list
        return final_list
    
    except TimeoutException :
        print('Cant search')
        return []
```

```python
food_list = create_list(driver,'food')
for i in food_list :
    print(i)
```

## Question 4 : Follow/Unfollow
> Follow/Unfollow given handle : 
    1. Open the Instagram Handle of ‚ÄúSo Delhi‚Äù
    2. Start following it. Print a message if you are already following.
    3. After following, unfollow the instagram handle. Print a message if you have already unfollowed.
    
- Make sure the page is loaded before calling the follow/unfollow function

```python
#Following functions requires the page to be loaded beforehand
def follow(driver):
    try :
        #locating the button
        btn = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_5f5mN')))
        
        #if the button is showing option to follow, click it
        if btn.text == 'Follow' : 
            btn.click()
            time.sleep(2) #required as it takes sometime after clicking the button
            
        #else either you are already following the handle or already requested to follow
        else : 
            print('Already following')
    
    #if unable to locate the button
    except TimeoutException :
        print("Cant follow")
 
def unfollow(driver):
    try :
        #locating the button
        btn = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_5f5mN')))

        #if the button is showing option to follow, click it
        if btn.text == 'Following' : 
            btn.click()

            #A confirmation dialogue is opened, click 'unfollow'
            pop_up = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'aOOlW')))
            pop_up.click()
            time.sleep(2) #required as it takes sometime after clicking the button


        #Already not following
        else : 
            print('Already not following')

    #if unable to locate the button
    except TimeoutException :
        print("Cant follow")
```

```python
follow(driver)
```

```python
unfollow(driver)
```

## Question 5 : Like/Unlike
> Like/Unlike posts :
    1. Liking the top 30 posts of the ‚Äòdilsefoodie'. Print message if you have already liked it.
    2. Unliking the top 30 posts of the ‚Äòdilsefoodie‚Äô. Print message if you have already unliked it.
    
- Make sure the page is loaded before calling the like_posts/unlike_posts function.
- increase the default wait time if slow internet
- Make sure the count(No. of posts to be likes/unliked) is less than total number of posts on page (Although this condition is handled by both functions)
- If a post is already liked/unliked the messege is printed along with the post number.
- The function closes the dialogue box in th end.

```python
#Following functions requires the page to be loaded beforehand

def like_posts(driver, count=1) :
    
    #try opening the first post
    try :
        li = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'v1Nh3')))
        li.click()
    except Exception as e:
        print("Cant Open Post") 
        return
    
    #running the loop for 'count' number of times, each iteration likes one post
    for i in range(count) :
        
        #Checking if the post is loaded by locating the dilogue box of post
        try : 
            temp = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'PdwC2')))
        except Exception as e:
            print("Loading error")
            return
        
        try :
            
            #checking if the post is already liked by locating the red heart button
            if len(driver.find_elements_by_class_name("glyphsSpriteHeart__filled__24__red_5")) != 0 :
                print("Post number ",i+1," is already liked")
            
            #else locating the white like button and clicking it
            else :
                btn = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'glyphsSpriteHeart__outline__24__grey_9')))
                btn.click()
        
        #neither red nor white button could be found
        except Exception as e:
            print("No like button")
       
        #clicking on next button
        try : 
            nxt = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'coreSpriteRightPaginationArrow')))
            nxt.click()
        
        #return from function if there are no more posts. Will execute if count>total number of posts
        except Exception as e:
            print("No more posts")
            return
    
    #closing the dialogue box
    driver.find_element_by_class_name('ckWGn').click()
    
        
def unlike_posts(driver, count=1) :
    
    #try opening the first post
    try :
        li = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'v1Nh3')))
        li.click()
    except Exception as e:
        print("Cant Open Post") 
        return
    
    #running the loop for 'count' number of times, each iteration likes one post
    for i in range(count) :
        
        #Checking if the post is loaded by locating the dilogue box of post
        try : 
            temp = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'PdwC2')))
        except Exception as e:
            print("Loading error")
            return
      
        #checking if the post is already unliked by locating the white like button
        if len(driver.find_elements_by_class_name("glyphsSpriteHeart__outline__24__grey_9")) != 0 :
            print("Post number ",i+1," is already unliked")
        
        #else locating the red like button and clicking it
        else :
            btn = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'glyphsSpriteHeart__filled__24__red_5')))
            btn.click()
       
        #clicking on next button
        try : 
            nxt = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'coreSpriteRightPaginationArrow')))
            nxt.click()
        
        #return from function if there are no more posts. Will execute if count>total number of posts
        except Exception as e:
            print("No more posts")
            return
    
     #closing the dialogue box
    driver.find_element_by_class_name('ckWGn').click()
```

## Question 6 : Extract list of followers
 > 1. Extract the usernames of the first 500 followers of ‚Äòfoodtalkindia‚Äô and ‚Äòsodelhi‚Äô.
2. Now print all the followers of ‚Äúfoodtalkindia‚Äù that you are following but those who don‚Äôt follow you.

- Although I tried to cover all cases but still if the function fails once, try running it again. 
- May not work properly if system is lagging or network is too slow
- Argument passed to Javascript code is the dialogue box i.e. followers panel.
- Followers are added to list one by one
- <font color=red> While scrolling down for loading more followers sometimes it get stuck at "See all suggestions", at that point, We need to scroll back up to where the suggestions have started. After that, followers start loading normally </font>
- It is assumed that count (Number of followers to be added) is less than total number of followers, otherwise function won't work.

#### Process used to create followers list :
1. Click on the followers button.
2. Locate the followers panel.
3. Start iterating all the followers one by one until no more are there.
4. Go to the bottom to load more followers and go to step 3 till required number of followers are added.
5. Close followers panel.

#### Process used to create my followers list :
1. My account is loaded by clicking on profile button
2. Total number of followers is calculated
3. create_followers_list function is called and total number o ffollowers is passed as an argument to it.

```python
#Following functions requires the page to be loaded beforehand

def create_followers_list(driver, count) :
    try :
        #locating the followers button and clicking it
        temp = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,'g47SY')))
        temp[1].click()
        
        # wait if followers are not loaded immidiately after clicking the followers button
        if len(driver.find_elements_by_class_name("wo9IH")) == 0 :
            time.sleep(3)
                  
    except Exception as e:
        print('Cant access followers list')
    
    try :
        #locate the followers panel
        followers_panel = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'isgrP')))
        
        i =0
        li =[]
        
        #add one follower in each iteration
        while i < count:
            try:
                #accessing ith follower
                follower = driver.find_elements_by_class_name('wo9IH')[i]
                li.append(follower.text) #append the text of webelement to list
                i += 1
            
            #if ith element is not loaded/located, go to the bottom of the list to load more
            except Exception as e:           
                #go to the bottom and wait for one second for more followers to load 
                driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight",followers_panel)  
                time.sleep(1)
                
                #if suggestions are shown at the bottom then more followers will not load
                if len(driver.find_elements_by_link_text("See All Suggestions")) : 
                    #go to the start of suggestions then automatically the followers will start loading
                    element = driver.find_element_by_class_name('vy6Bb')
                    element.location_once_scrolled_into_view
                    #driver.execute_script('arguments[0].scrollBy(0, -100);',followers_panel)
                    time.sleep(2)

                                   
        #close the followers panel after the loop terminates
        driver.find_element_by_class_name("glyphsSpriteX__outline__24__grey_9").click()
        return li
    
    except Exception as e:
        print('Cant access followers')
        return []
    
    
    
def my_followers_list(driver) :
    
    #open your instagram handle by clcking on profile button    
    try :
        profile_btn = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,'XrOey')))[-1].click()
    except Exception as e:
        print('Cant open profile')
        return []
    
    try :
        #get the number of followers
        temp = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,'g47SY')))
        count = int(temp[1].get_attribute('title').replace(',',''))
        
    except Exception as e:
        print('Cant access followers list')
    
    # call create_followers_list function and pass the count
    return create_followers_list(driver,count)
        
    
```

## Question 7 : Check story
 > Check the story of ‚Äòcoding.ninjas‚Äô. Consider the following Scenarios and print error messages accordingly -
1. If You have already seen the story.
2. Or The user has no story.
3. Or View the story if not yet seen.

- Make sure the page is loaded beforehand
- An extra function to close the story is also provided.

```python
#Following functions requires the page to be loaded beforehand
def view_story(driver) :
    try :
        #locating the the profile pic to click to view story
        a = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_2dbep'))) 
        b = a.find_element_by_xpath('..') #b is parent of a in html doc
        
        #getting html content of b
        data = BeautifulSoup(b.get_attribute('outerHTML'), 'html.parser') 
            
            
        '''     b is two attributes of our use,
                   if b/canvase.height is 174 or 332 then story is already seen
                   elseif b/div has one class then no story is availabe
                   else the story can be viewed
        '''
        
        height = int(data.canvas['height']) 
        siz = len(data.div['class'])
        
        if height == 174 or height == 332 :
            print('Story already seen!')
            return
        
        elif siz == 1 :
            print('No story available!')
            return
        
        else :
            a.click()
            
            
    except Exception as e:
        print('Cant access story')

        
def close_story(driver) :  #if you want to close story earlier 
    try :
        driver.find_element_by_class_name("Szr5J").click()
    except Exception as e:
        print('Cant close')
```

```python
search_and_open(driver, 'coding ninjas')
time.sleep(2)
view_story(driver)
time.sleep(3)
close_story(driver)
```
